### Random tensors

Why random tensors?

Random tensors are important because the way many neural networks learn is that they start with tensors full of random numbers and then adjust to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(244,244,3)) # height, width, colour channel
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Tensor datatypes

1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on right device

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0,6.0,9.0], dtype=None, device = None, requires_grad = False)
# dtype = what datatype the tensor is
# device = what device is your tensor one
# requires_grad = whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor * float_32_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9], dtype=torch.long)
int_32_tensor
```

```python
float_32_tensor * int_32_tensor
```

### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2], [3,4], [5,6]])
tensor_B = torch.tensor([[7,10],[8,11],[9,12]])

torch.mm(tensor_A, tensor_B) #torch.mm is the same as torch.matmul
```

```python
tensor_A.shape, tensor_B.shape
```

To fix tensor shape issues, we can manipulate the shape of one of our tensors using trarnspose

A **transpose** switches the axes of dimensions of a given tensor

```python
tensor_B.T, tensor_B
```

```python
torch.mm(tensor_A, tensor_B.T)
```

## Reshaping, stacking, squeeing, and unsqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - remove all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (wrapped) in a certain way

```python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original)
z[:,0] = 5
z,x
```
